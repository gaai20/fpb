```
import numpy as np
```

```
w0=np.array([[1,1,1],
            [0,0,0],
            [2,2,2]])
```

```
w1=np.array([[1,1,1],
             [1,1,1],
             [0,0,0]])
```

```
te=w0.dot(w1)
te
```

```
def neu_net(inputs,weights):
  hid=inputs.dot(weights[0])
  pred=hid.dot(weights[1])
  return pred
```

```
weights=np.array([w0,w1])
toes=np.array([1,1,1])
wins=np.array([1,1,1])
nfans=np.array([0,0,0])
inputs=np.array([toes[0],wins[0],nfans[0]])
```

```
pred=neu_net(inputs,weights)
print(pred)
```
